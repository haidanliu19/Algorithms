```python
N = int(input())

arrays = []

for _ in range(N) : 
    arrays.append(int(input()))

N = int(input())

arrays = []

m = 10001
count = [0 for i in range(m)]

for i in range(len(arrays)):
    count[arrays[i]-1] += 1


for i in range(m):
    if (count[i] !=0):
        for j in range(0, count[i]):
            print(i+1)
```

```python
import sys

N = int(sys.stdin.readline())

m = 10001
count = [0 for i in range(m)]

for _ in range(N) : 
    num = int(sys.stdin.readline())
    count[num] += 1

for i in range(m):
    for j in range(0, count[i]):
        print(i)
```

```python
N = int(input())

arrays = []

for _ in range(N) : 
    arrays.append(int(input()))

N = int(input())

arrays = []

m = 10001
count = [0 for i in range(m)]

for i in range(len(arrays)):
    count[arrays[i]-1] += 1

for i in range(m):
    while (count[i] !=0):
        print(i+1)
        count[i] -= 1
```
