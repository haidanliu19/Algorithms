```python
def dp(x):
    # 이 부분 없으면 무한 반복
    if (x == 1) or (x == 2):
        return 1
    # 자기 앞의 숫자와 자기 앞의 앞의 숫자
    return dp(x-1) + dp(x-2)
    
x = 10
dp(x)
```

```python
d = [0]*100

def dp(x):
    # 이 부분 없으면 무한 반복
    if (x == 1) or (x == 2):
        return 1
    if (d[x]!=0):
        return d[x]
    # 자기 앞의 숫자와 자기 앞의 앞의 숫자
    #return d[x] = dp(x-1) + dp(x-2)
    d[x] = dp(x-1) + dp(x-2)
    return d[x]
    
x = 50
dp(x)
```
