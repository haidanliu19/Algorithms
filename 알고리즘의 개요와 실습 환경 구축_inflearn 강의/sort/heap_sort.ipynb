```python
%%timeit

heap = [7,6,5,8,3,5,9,1,6]
number = len(heap)
def heap_sort(heap, number):
    # 먼저 전체 트리 구조를 힙구조로 바꿉니다.
    for i in range(number):
        c = i
        while  (c != 0):
            root = int((c-1) /2)
            if (heap[root] < heap[c]): # 부모의 값보다 자신의 값이 더 크다면 
                heap[root], heap[c] = heap[c], heap[root]
            c = root # 자식의 부모로 이동해서 반복적으로 수행

    
    # 크기를 줄여가며 반복적으로 힙을 구성
    for i in range(number-1, 0, -1):
        #가장 큰 값을 맨 뒤어로 보낸다.
        heap[i], heap[0] = heap[0] , heap[i]
        root = 0
        c = 1
        #힙 구조를 만들어주는 부분
        while (c < i):
            c = 2 * root + 1 # root 자식
            # 자식 중에 더 큰 값을 찾기
            if (c < (i-1)) and (heap[c] < heap[c+1]) :
                c+=1
            if (c <i) and (heap[root] < heap[c]):
                heap[root], heap[c] = heap[c], heap[root]
            root = c

    return heap

heap_1 = heap_sort(heap, number)
print(heap_1)
```
