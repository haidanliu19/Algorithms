```python
%%time

def solution(s):
    answer = "".join(sorted(s,reverse=True))
    return answer

s = "Zbcdefg"
print(solution(s))
```

```python
%%time

def solution(s):
    answer = "".join(sorted(s)[::-1])
    return answer

s = "Zbcdefg"
print(solution(s))
```
